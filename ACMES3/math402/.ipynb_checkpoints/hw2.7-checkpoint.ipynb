```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn
from scipy.stats import chi2
import scipy.integrate as integrate
from random import randint
%matplotlib inline
```

### 5.1 

### (A)

```python
k_list = [2,4,6]
for k in k_list:
    n = np.float(10**k)
    sn_points = np.random.randn(n)
    z_points = sn_points**2
    plt.hist(z_points, bins = 50)
    plt.show()
    x_list = [0.5,1.0,1.5]
    for x in x_list:
        print "CDF = {} \t for x = {}".format(np.sum(z_points < x)/n, x)
        
    print "E[x] = {}".format(np.mean(z_points))
    print "V[x] = {}".format(np.mean(z_points**2) - np.mean(z_points)**2)
    
```

### (B)

```python
k_list = [2,4,6]
for k in k_list:
    n = np.float(10**k)
    z_points = chi2.rvs(1,size = n)
    plt.hist(z_points, bins = 50)
    plt.show()
x_list = [0.5,1.0,1.5]
for x in x_list:
    print "CDF = {} \t for x = {}".format(chi2.cdf(x, 1), x)

print "E[x] = {}".format(chi2.stats(1, moments = 'm'))
print "V[x] = {}".format(chi2.stats(1, moments = 'v'))
```

### 5.3

```python
k_list = [2,4,6]
y = lambda x: np.exp(np.cos(x**2))
for k in k_list:
    n = np.float(10**k)
    x_points = np.random.uniform(0,1,n)
    y_points = np.random.uniform(0,3,n)
    print np.sum(y_points < y(x_points))/n * 3*1
    
print 'Actual result:',integrate.quad(y, 0, 1)[0]
```

I just used a number that gave an answer that was close enough to satisfy a relatively small error tolerance.

```python
def roll(ndice,nsides=4):
    return [randint(1,nsides) for _ in range(ndice)]

def count_it():
    c = roll(10)
    ones = c.count(1)
    twos = c.count(2)
    threes = c.count(3)
    fours = c.count(4)
    return (ones + twos + threes * (2) + fours *(-1)) < 0

k_list = [5]
for k in k_list:
    rolls = float(10**k)
    print (sum(1 for _ in range(int(rolls)) if count_it())/rolls)
```

$10^{5}$ gives a satisfactorily close answer.
