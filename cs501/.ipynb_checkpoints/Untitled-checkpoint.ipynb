```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from IPython.display import Image


%matplotlib inline

WIDTH = 512
HEIGHT = 288

rand_array=np.random.rand(HEIGHT, WIDTH) #create your array
plt.imshow(rand_array,cmap=cm.bone) #show your array with the selected colour
plt.show() #show the image

for i in range(100):
    plt.plot(np.linspace(np.random.randint(0,HEIGHT), np.random.randint(0,HEIGHT), WIDTH))
plt.show()

plt.scatter(np.random.randint(0,WIDTH,100),np.random.randint(0,HEIGHT,100))
plt.show()

rand_array=np.random.rand(HEIGHT, WIDTH) #create your array
plt.imshow(rand_array,cmap=cm.cool) #show your array with the selected colour
plt.show()

rand_array=np.random.rand(HEIGHT, WIDTH) #create your array
plt.imshow(rand_array,cmap=cm.seismic) #show your array with the selected colour
plt.show()
```
