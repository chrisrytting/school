```python
def unpickle(file):
    import cPickle
    fo = open(file, 'rb')
    dict = cPickle.load(fo)
    fo.close()
    return dict
```

```python
import numpy as np
import pandas
from matplotlib import pyplot as plt
import sys

%matplotlib inline
```

```python
data = pandas.read_csv( 'Fisher.csv' )
m = data.as_matrix()
labels = m[:,0]
labels[ labels==2 ] = 1  # squash class 2 into class 1
labels = np.atleast_2d( labels ).T
features = m[:,1:5]
```

```python
class Perceptron:
    def __init__(self, data, labels, step_size):
        self.data = data
        self.data = np.hstack((self.data, np.ones((self.data.shape[0], 1))))
        self.n_obs = np.shape(self.data)[0]
        self.n_feat = np.shape(self.data)[1]
        self.labels = labels.flatten()
        self.c = step_size
        self.weights = np.random.normal(0,.001,self.n_feat)
        self.pred = np.zeros(self.n_obs)
        self.class_acc = []
        self.l2_wghts = []
    def f(self, x):
        if np.dot(self.weights,x) > 0:
            return 1
        else:
            return 0
        
    def train(self, iters = 100):
        for i in xrange(iters):
            for j in xrange(self.n_obs):
                y = self.f(self.data[j])
                self.pred[j] = y
                self.weights = self.weights + self.c*(self.labels[j] - y)*self.data[j]
            
            self.class_acc.append((float(len(self.pred)) - np.sum(np.abs(self.pred - self.labels)))/float(len(self.pred)))
            norm = np.sqrt(np.dot(self.weights, self.weights))
            self.l2_wghts.append(norm)
           
```

```python
p1 = Perceptron(features, labels, 1.0)
p1.train()
print p1.weights

p2 = Perceptron(features, labels, 1.0)
p2.train()
print p2.weights

p3 = Perceptron(features, labels, 1.0)
p3.train()
print p3.weights
```

```python
plt.plot(p1.class_acc, label = 'c=1e1')
plt.plot(p2.class_acc, label = 'c=1e-1')
plt.plot(p3.class_acc, label = 'c=1e-2')
plt.xlabel('Iterations')
plt.ylabel('Classification Accuracy')
plt.legend(loc=0)
plt.show()

t = np.linspace(1,100,100)


plt.plot(t,np.log(p1.l2_wghts), label='c=1')
plt.plot(t,np.log(p2.l2_wghts), label='c=0.1')
plt.plot(t,np.log(p3.l2_wghts), label='c=0.01')
plt.xlabel('Iterations')
plt.ylabel('Natural Log of L2 Norm of Weights')
plt.legend(loc=0)
plt.show()
```

```python
def unpickle( file ):
    import cPickle
    fo = open(file, 'rb')
    dict = cPickle.load(fo)
    fo.close()
    return dict
 
data2 = unpickle( 'cifar-10-batches-py/data_batch_1' )
 
features2 = data2['data']
labels2 = data2['labels']
labels2 = np.atleast_2d( labels2 ).T
 
# squash classes 0-4 into class 0, and squash classes 5-9 into class 1
labels2[ labels2 < 5 ] = 0
labels2[ labels2 >= 5 ] = 1
```

```python
p4 = Perceptron(features2, labels2, 0.001) 
p4.train()
print p4.weights

p5 = Perceptron(features2, labels2, 0.00001) 
p5.train()
print p5.weights

plt.plot(t, p4.class_acc, label='c=0.001')
plt.plot(t, p5.class_acc, label='c=0.00001')
plt.xlabel('Iterations')
plt.ylabel('Classification Accuracy')
plt.legend(loc=0)
plt.show()

plt.plot(t, np.log(p4.l2_wghts), label='c=0.001')
plt.plot(t, np.log(p5.l2_wghts), label='c=0.00001')
plt.xlabel('Iterations')
plt.ylabel('Natural Logarithm of L2 Norm of Weight Vector')
plt.legend(loc=0)
plt.show()
```
